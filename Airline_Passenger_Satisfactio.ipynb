```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
fig.suptitle('Tỷ lệ Hài lòng trong từng nhóm Khách hàng', fontsize=14)

# Dữ liệu nhóm Loyal
loyal = data[data['Customer Type'] == 'Loyal Customer']
loyal_counts = loyal['satisfaction'].value_counts()

axs[0].pie(
    loyal_counts,
    labels=loyal_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
axs[0].set_title('Loyal Customer')

# Dữ liệu nhóm Disloyal
disloyal = data[data['Customer Type'] == 'disloyal Customer']
disloyal_counts = disloyal['satisfaction'].value_counts()

axs[1].pie(
    disloyal_counts,
    labels=disloyal_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
axs[1].set_title('Disloyal Customer')

plt.tight_layout()
plt.show()
```

**Nhóm Khách hàng trung thành (Loyal Customer):** 48.1% cảm thấy hài lòng, 51.9% cảm thấy trung lập hoặc không hài lòng.

Gần một nửa trong số nhóm trung thành không hài lòng hoặc trung lập. Như vậy, khách hàng trung thành không đồng nghĩa với việc họ hài lòng về dịch vụ trong các chuyến bay. Họ gắn bó với hàng hàng không do một số lý do khác như thói quen, các chương trình tích điểm, hội viên, ít lựa chọn thay thế, giá cả ổn định, thời gian phù hợp,...

**Nhóm Khách hàng không trung thành (Disloyal Customer):** Chỉ 25.2% hài lòng và 74.8% trung lập hoặc không hài lòng.

Đa số khách hàng không trung thành cũng không hài lòng. Mức độ hài lòng thấp đã giải thích lý do vì sao họ không gắn bó. Mức độ hài lòng thấp cũng cho thấy dịch vụ của hãng đang không tốt.

```python
data['Total delay'] = data['Departure Delay in Minutes'] + data['Arrival Delay in Minutes']
delay = data.groupby('Class')['Total delay'].mean().sort_values(ascending=False)
delay
```

```python
delay.plot(kind='bar',color=['green', 'coral', 'steelblue'], figsize=(7,4))
plt.title('Trung bình tổng thời gian trễ theo hạng vé')
plt.ylabel('Tổng thời gian trễ trung bình (phút)')
plt.xlabel('Class')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

- Vé Eco có thời gian trễ trung bình cao nhất, khoảng 30 phút.

- Vé Eco Plus đứng thứ hai, với thời gian trễ thấp hơn một chút.

- Vé Business có thời gian trễ thấp nhất trong ba loại.

Tuy nhiên, thời gian trễ trung bình giữa các loại vé không có sự chênh lệch lớn, chỉ khoảng 2 phút.

```python
bins = [0, 30, 60, 90, 120, 150]
labels = ['(0-30]', '(30-60]', '(60-90]', '(90-120]', '(120-150]']
data['Delay Bin'] = pd.cut(data['Total delay'], bins=bins, labels=labels)

# Tính tỷ lệ % theo từng nhóm delay bin
grouped = data.groupby(['Delay Bin', 'satisfaction']).size().reset_index(name='count')
total_per_bin = grouped.groupby('Delay Bin')['count'].transform('sum')
grouped['percentage'] = grouped['count'] / total_per_bin * 100

pivot_data = grouped.pivot(index='Delay Bin', columns='satisfaction', values='percentage').fillna(0)
pivot_data = pivot_data[['satisfied', 'neutral or dissatisfied']]

# Vẽ biểu đồ stacked bar
pivot_data.plot(kind='bar', stacked=True, color=['#4c72b0', '#dd8452'], figsize=(8,5))
plt.ylabel('Tỷ lệ (%)')
plt.title('Tỷ lệ % khách hàng hài lòng theo từng mức độ trễ')
plt.legend(title='')
plt.tight_layout()
plt.show()
```

Khi thời gian trễ tăng, tỷ lệ khách hàng hài lòng giảm dần rõ rệt.

- Thời gian trễ càng tăng làm cho mức độ hài lòng có xu hướng giảm.

- Tuy nhiên, từ mức trễ cao trở lên (~60 phút), tỷ lệ hài lòng ổn định ở mức thấp, cho thấy khách hàng có thể đã quen với trải nghiệm kém và không kỳ vọng nhiều hơn.

Hãng bay nên tập trung cải thiện trải nghiệm ở những chuyến trễ vừa phải (30–90 phút), vì đó là vùng chuyển hóa cảm xúc rõ nhất của khách.
